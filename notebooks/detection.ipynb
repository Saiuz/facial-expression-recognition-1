```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

%matplotlib inline
```

### Read and prepare the image

```python
image = cv.imread('../data/original/test2.jpg')

image_bw = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(image_bw, cmap='gray')
plt.axis('off')
```

```python
def convertToRGB(image):
    return cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

### Load pretrained model

For baseline of our project, we decided to use pretrained Haar Cascade model for front face detection.

Since we don't have a dataset on which we can asses both fact detection and facial recognition model performance, we would not concentrate on the implementation of this part.

But, in general, to make our pipeline more homogeneous, in the next iteration we are going to implement Haar Cascade by ourselves.

```python
haar_cascade_face = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(image_bw, scaleFactor = 1.2, minNeighbors = 5);
```

```python
print('Faces found: ', len(faces_rects))
```

```python
faces = []

for (x,y,w,h) in faces_rects:
    faces.append(image_bw[y : y + h, x : x + w])
```

```python
len(faces)
```

```python
faces = [face for face in faces if face.shape[0] * face.shape[1] != 0]
```

```python
fig = plt.figure(figsize=(13,13))

rows = 6
cols = 6

for i in range(len(faces)):
    img = faces[i]
    fig.add_subplot(rows, cols, i+1)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
plt.show()
```

```python
for i in range(len(faces)):
        resized = cv.resize(faces[i], (100, 100))
        plt.imsave('/home/dinamic/Desktop/faces/face_' + str(i) + '.png', resized, cmap = 'gray')
```

### Plot boxes on detected faces

```python
for (x,y,w,h) in faces_rects:
     cv.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(convertToRGB(image))
plt.axis('off')
```

On the whole, this pretrained model shows pretty descent results. Thre are few faces that this model fails to capture, especially if we increase the size of the bounding box or `minNeighbors` value.
