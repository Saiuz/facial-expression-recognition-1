```python
import numpy as np
import cv2 as cv

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image = cv.imread('../data/test2.jpg')

image_bw = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_bw, cmap='gray')
```

```python
def convertToRGB(image):
    return cv.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(image_bw, scaleFactor = 1.2, minNeighbors = 5);
```

```python
print('Faces found: ', len(faces_rects))
```

```python
faces = []

for (x,y,w,h) in faces_rects:
    faces.append(image_bw[y - h//2 : y + h + h//2, x - w//2 : x + w + w//2])
     #cv.rectangle(image, (x - 40, y - 40), (x+w+40, y+h+40), (0, 255, 0), 2)
```

```python
len(faces)
```

```python
faces = [face for face in faces if face.shape[0] * face.shape[1] != 0]
```

```python
fig = plt.figure(figsize=(13,13))

rows = 6
cols = 6

for i in range(len(faces)):
    img = faces[i]
    fig.add_subplot(rows, cols, i+1)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
plt.show()
```

```python
for i in range(len(faces)):
    
        resized = cv.resize(faces[i], (100, 100))
        plt.imsave('/home/dinamic/Desktop/faces/face_' + str(i) + '.png', resized, cmap = 'gray')
```
