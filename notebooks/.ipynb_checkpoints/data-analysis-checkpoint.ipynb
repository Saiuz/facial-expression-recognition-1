## Analysis of the `fer2013` dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

The data consists of `48x48` pixel grayscale images of faces. The faces have been automatically registered so that the face is more or less centered and occupies about the same amount of space in each image. The task is to categorize each face based on the emotion shown in the facial expression in to one of seven categories (`0=Angry`, `1=Disgust`, `2=Fear`, `3=Happy`, `4=Sad`, `5=Surprise`, `6=Neutral`).

```python
mapping = {
    0: 'Angry', 
    1: 'Disgust', 
    2: 'Fear', 
    3: 'Happy', 
    4: 'Sad', 
    5: 'Surprise', 
    6: 'Neutral'   
}
```

### Modify original dataset format

The dataset has all pixels merged into one field in `.cvs` file which are space-separated. That makes it highly uncomfortable to work with it.

```python
index = [i for i in range(35887)]

columns = ['y'] + [str(i) for i in range(48 * 48)] + ['category']
```

```python
df = pd.DataFrame(columns=columns, index = index)
```

```python
with open('../data/emotions/fer2013/fer2013.csv') as file:
    file.readline()
    
    for i in range(35887):
        print(i)
        line = file.readline()
        
        y, pixels, category = line.split(',')
        
        pixels = [p for p in pixels.split(' ')]
        
        df.at[i, 'y'] = y
        
        for index, value in enumerate(pixels):
            df.at[i, str(index)] = value
            
        df.at[i, 'category'] = category
```

```python
df.to_csv('../data/emotions/emotions.csv')
```

#### Distribution of classes in our dataset

```python
stats = df.groupby(['category', 'y']).aggregate({'y': ['count']})

stats.columns = ["_".join(x) for x in stats.columns.ravel()]
```

```python
stats = stats.reset_index().pivot_table(index='y', values='y_count', columns='category').reset_index()
```

```python
stats['% of all'] = ((stats['PrivateTest'] + stats['PublicTest'] + stats['Training']) / len(df) * 100) 
```

```python
stats
```

It is obvious that classes in the dataset are not well-balanced. First of all, we have highly underepresented `Disgust` class, which will for sure need some special treatment. Only `1.5%` of samples correspond to this class. 

Regarding other classes, we still have a huge difference between some of them. For example, `25%` of samples correspond to the `Happy` class, which is more than the fraction of `Angry` and `Surprise` altogether (`24.9%`). 

It may seem as not big deal, but, in general, it will not only cause the model to predict overrepresented classes more accurately, it will also cause the model to predict a big number of false positives. 

There are few possible scenarios how can we tackle this issue:

    1) Try to augment underepresented classes using standard image augmentation techniques (it will make training    more difficult, since the size of the training set will increase). 

    2) Try to use other performance measures. For example, asses the peformance of the model on seperate classes.
    
    3) Delete instances from over-represented classes, such as Happly, Sad and Fear (simple procedure and easy to implement).
    
    4) Try different alogrithm (not really the case when we are working with image data)

#### Image quality

It is considered that the human accuracy on the given dataset is `65.5%` ([source](https://arxiv.org/pdf/1612.02903.pdf)). That number seems a bit low, if we think of our ability to recognize and identify emotions as one of the most important elements of human communication. The reason for this is that sometimes it's really hard to tell from `48x48` image what emotion is beeing depicted. It could be simply an unusual aspect or just ambigous expression that makes it hard even for humans to correctly classify emotions represented on the images. And of course images may be just labeled incorrectly, which makes it impossible to even guess the correct class.

Let's take look on 100 random samples and try to identify mislabeled images. Of course this approach is not representative, but at least we will get a general labeling quality pictures.

```python
np.random.seed(1)

mislabeled = {
    7752: "Neutral",
    35731: "Fear",
    22927: "!!!",
    31960: "Happy",
    33800: "Angry",
    2562: "Happy",
    432: "Neutral",
    16927: "Disgust",
    29994: "Neutral",
    27713: "Neutral",
    13682: "Sad",
    29237: "Neutral",
    9765: "Sad",
    16540: "!!!"
}

fig = plt.figure(figsize=(16, 16))

rows = 10
cols = 10

for i in range(1, rows * cols + 1):
    idx = np.random.randint(1, len(df))
    
    fig.add_subplot(rows, cols, i)
    image = np.array(df.iloc[idx, 1:2305].values, dtype=float)
    plt.imshow(image.reshape((48, 48)), cmap = 'gray')
    if idx in mislabeled.keys():
        plt.title(str(idx) + ": " + df.y[idx], fontdict = {'fontsize': 10, 'color': "red"})
    else:
        plt.title(str(idx) + ": " + df.y[idx], fontdict = {'fontsize': 10, 'color': "green"})
    plt.axis('off')
plt.show()
```

I would not argue that identification of human emotions is an objective thing. It's not. Different expressions mean different emotions at different parts of the world. There is a huge cultural difference in how people express and perceive emotions. 

The dataset we are working with was most probably createding using web crawling. It is diverse in age, race and gender. And this is a good point. The bad point is that it containes ralatively a number of trash images. For example, in our randome sample there are 2 images which don't directly depict a face (`22927` and `16540`). Besides it, a number of images have inaccurate labeling. We marked them with red text color in the grid above.

In internet, as in life in general, some kind of emotions are naturally underepresented. You can't usually see pronounced fear, anger or disgust on human face (at least in developed countries). And people often hide their emotions, which makes it even harder to determine the true feelings behind the facial expression.

Those factors make the problem of facial expression recognition even harder. And now it's not surprising that human score on the given dataset is lower that we would expect.

In general, there is no simple and fast way to deal with dataset problems discussed above. We can try to train a separate classifier for face detection and remove non-face images. 

Additionally we can try to manually relabel incorrectly labeled images. But this approach, as any manual work, is prone to mistakes and bias regarding the "true" emotion expresion. 

#### Visualization of the dataset (subsample) using $t$-SNE algorithm

One of the ways to determine whether a classification task is easy or hard for a computer is to try to visualize the images in 2-dimentional space. For this purpose $t$-SNE algorithms have been developed. It has been proven that this visualization algorithm provides provides the greatest capabilities in comparison with other similar techniques and methods.

```python
from sklearn.manifold import TSNE

from matplotlib.offsetbox import OffsetImage, AnnotationBbox

import time
```

```python
colormaps = {
    'Angry': 'magma', 
    'Disgust': 'pink', 
    'Fear': 'winter', 
    'Happy': 'spring', 
    'Sad': 'autumn', 
    'Surprise': 'summer', 
    'Neutral': 'gray'  
}
```

```python
def imscatter(x, y, images, labels, ax=None, zoom=1):        
    x, y = np.atleast_1d(x, y)
    for x0, y0, i0, l0 in zip(x, y, images, labels):
        ab = AnnotationBbox(OffsetImage(i0, zoom = zoom, cmap = colormaps[l0]), 
                            (x0, y0),
                            xycoords='data', 
                            frameon=False)
        ax.add_artist(ab)
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()
```

```python
sample = df.sample(500)
```

```python
X = np.array(sample.iloc[:, 1:-1])

labels = sample.y
```

```python
images = X.reshape((len(X), 48, 48))
```

```python
start = time.time()
points = TSNE(n_components=2, perplexity=5, verbose=1).fit_transform(X)
end = time.time()
print('\n')
print(end - start)
print('\n')
```

On the following visualization you observe 500 random images from `fer2013` dataset, which are shown in different colormaps according to their label.

```python
x = points[:, 0]
y = points[:, 1]
    
fig, ax = plt.subplots(figsize = (18,18))
imscatter(x, y, images, labels, zoom=0.65, ax=ax)
    
ax.scatter(x, y)

plt.savefig('../data/t-sne-visualization.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

As we can observe $t$-SNE algorithm doesn't perform well with this kind of data. It does a great job of identifying  age, gender, race or skin colour patterns, but fails to correctly construct facial expression clusters. This means, that, in fact, recognition of facial expression is not an easy task for a machine algorithm to do, because visual difference could be really small for images of absolutely different emotions.
